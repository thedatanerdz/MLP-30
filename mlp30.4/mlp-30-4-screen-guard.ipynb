## **Contingency table**

A contingency table is a way to organize information or data in a table format.
It is like a chart that shows how things are related or connected to each other.
Imagine you have two categories or groups, like boys and girls. A contingency table can show how many boys and how many girls there are.
Each category or group is shown as a row or a column in the table.
The table has numbers that tell us how many things belong to each combination of categories.
For example, the table can show how many boys like basketball and how many girls like basketball.
Contingency tables are helpful because they make it easier to see patterns or relationships between different categories of things.
By looking at the numbers in the table, we can understand how different categories are connected or related to each othe

```python
CT_values = pd.crosstab(df["suggestion_type"], df["purchase_flag"] ).values
CT_values 
```

```python
CT_values[1,1]
```

```python
CT_values[1,0]
```

```python
CT_values[0,0]
```

```python
CT_values[0,1]
```

```python
# purchase rate
with_phone_rate = CT_values[1,1] / (CT_values[1,1] + CT_values[1,0]) 
with_cover_rate = CT_values[0,1] / (CT_values[0,0] + CT_values[0,1])
print(with_phone_rate, with_cover_rate )
```

31% phone witn no cover 
39% phone with cover

null hypothesis = no relationship between purchase rate and recommendation type

```python
from scipy.stats import chi2_contingency

chi2_statistic, p_value, dof, expected_values = chi2_contingency(CT_values, correction=False)
print(chi2_statistic, p_value)
```

```python
prob = 0.95
alpha = 1.0 - prob
print('significance=%.3f, p=%.3f' % (alpha, p_value))
if p_value <= alpha:
    print('Dependant we can reject H0 safely')
else:
     print('Independant we cannot reject H0 safely')
```
